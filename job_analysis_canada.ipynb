```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the dataset
file_path = 'C:/Users/study/Downloads/vscode/Cleaned_Dataset.csv'
job_data = pd.read_csv(file_path)
```

```python
job_data.head(5)
```

```python
# Get basic information about the dataset
job_data.info()
```

```python
# 0 nul values
# data types also correct 
job_data.isnull().sum()
```

```python
# Job Titles and Roles in Demand
top_job_titles = job_data['Job Title'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_job_titles.index, x=top_job_titles.values, palette="Blues_d")
plt.title("Top 10 Most Common Data Analyst Job Titles")
plt.xlabel("Number of Jobs")
plt.ylabel("Job Title")
plt.show()
```

```python
# Top Skills Required
top_skills = job_data['Skill'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_skills.index, x=top_skills.values, palette="Greens_d")
plt.title("Top 10 Most In-Demand Skills")
plt.xlabel("Number of Jobs")
plt.ylabel("Skill")
plt.show()
```

```python
# Job Opportunities by Province and City
plt.figure(figsize=(10, 6))
city_job_counts = job_data['City'].value_counts().nlargest(10)
sns.barplot(y=city_job_counts.index, x=city_job_counts.values, palette="Purples_d")
plt.title("Top 10 Cities with Most Data Analyst Job Listings")
plt.xlabel("Number of Jobs")
plt.ylabel("City")
plt.show()

plt.figure(figsize=(10, 6))
province_job_counts = job_data['Province'].value_counts().nlargest(10)
sns.barplot(y=province_job_counts.index, x=province_job_counts.values, palette="Reds_d")
plt.title("Top 10 Provinces with Most Data Analyst Job Listings")
plt.xlabel("Number of Jobs")
plt.ylabel("Province")
plt.show()
```

```python
#  Employers and Industries Hiring the Most
top_employers = job_data['Employer'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_employers.index, x=top_employers.values, palette="Oranges_d")
plt.title("Top 10 Employers Hiring Data Analysts")
plt.xlabel("Number of Jobs")
plt.ylabel("Employer")
plt.show()
```

```python
#  Comparison of Entry-Level and Senior Positions
plt.figure(figsize=(8, 6))
seniority_counts = job_data['Seniority'].value_counts()
sns.barplot(x=seniority_counts.index, y=seniority_counts.values, palette="coolwarm")
plt.title("Comparison of Entry-Level vs. Senior Job Opportunities")
plt.xlabel("Seniority Level")
plt.ylabel("Number of Jobs")
plt.show()
```

```python
# Salary by Work Type and Seniority
plt.figure(figsize=(10, 6))
sns.boxplot(x='Work Type', y='Avg_Salary', data=job_data, palette="Set2")
plt.title("Salary Range by Work Type")
plt.xlabel("Work Type")
plt.ylabel("Average Salary (CAD)")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Seniority', y='Avg_Salary', data=job_data, palette="Set3")
plt.title("Salary Range by Seniority")
plt.xlabel("Seniority")
plt.ylabel("Average Salary (CAD)")
plt.show()
```
